```python
rel_2=pd.crosstab(df['Q6'],df['Q19'])
rel_2_normalize = rel_2.div(rel_2.sum(1).astype(float), axis=0)
rel_2_normalize.plot(kind='bar',title='Survival Rate by Passenger Classes',stacked=True,grid=True)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
df['Q21'].value_counts().plot(kind='pie',title="When people violate traffic often")
plt.show()
```

```python
rel_1=pd.crosstab(df['Q12'],df['Q19'])
rel_1_normalize = rel_1.div(rel_1.sum(1).astype(float), axis=0)
rel_1_normalize.plot(kind='bar',title='Accident Rate vs Time',stacked=True,grid=True)
plt.xlabel('Travel time')
plt.ylabel('Accident rate')
plt.show()
```

**The above plot clearly shows that most of the accidents are happened during night.**
